## Primero el DataSet

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
```

```python
#Creamos nuestros datos artificiales, donde se buscará clasificar dos anillos concéntricos de datos.
x, y = make_circles(n_samples=500, noise=0.05, factor=0.5)

# Aquí corregimos la variable "y" que viene a ser de la forma (500, )
#y = y[:, np.newaxis] #Y ahora lo convertimos en un vector como tal -> (500, 1)
#print(y.shape) #Para verificarlo podemos ejecutar esto.

#Resolución del mapa de predicción: (es decir, resolución del gráfico)
res=100

#Coordenadas del mapa de predicción:
_x0 = np.linspace(-1.5, 1.5, res)
_x1 = np.linspace(-1.5, 1.5, res)

#***************************************************************
#Esta sección no influye en el gráfico, es más, el gráfico se puede hacer sin esto.

#Input con cada combo de coordenadas del mapa de predicción:
_pX = np.array(np.meshgrid(_x0, _x1)).T.reshape(-1, 2) #No se usa para nada del gráfico.

#Objeto vacío a 0.5 del mapa de predicción:
_pY = np.zeros((res, res))+0.5 #Se usa más adelante para almacenar el resultado del vector de predicciones.

#Visualización del mapa de predicción:
plt.figure(figsize=(7, 7)) #Definimos el tamaño del gráfico: alto x ancho
plt.pcolormesh(_x0, _x1, _pY, cmap="coolwarm", vmin=0, vmax=1)

#***************************************************************

#Visualización de la nube de datos:
plt.scatter(x[y==0, 0], x[y==0, 1], c="red") #Con esto le decimos que solo dibuje uno de los circulos
#plt.scatter(x[y[:, 0]==0, 0], x[y[:, 0]==0, 1], c="red")
plt.scatter(x[y==1, 0], x[y==1, 1], c="orange")
#plt.scatter(x[y[:, 0]==1, 0], x[y[:, 0]==1, 1], c="orange")
plt.tick_params(labelbottom=False, labelleft=False) #Dejamos de mostrar los ejes.
plt.show()
```

## Usando TensorFlow v1

```python
from matplotlib import animation
from IPython.core.display import display, HTML

import tensorflow.compat.v1 as tf #Lo importamos pero como si fuese la v1 de TensorFlow
tf.disable_v2_behavior() #Con esto deshabilitamos el "eager".
#Advertencia: tensorflow:From C:\Users\Moises\anaconda3\envs\IA_ML_2_0\lib\site-packages\tensorflow_core\python\compat\v2_compat.py:88: disable_resource_variables (from tensorflow.python.ops.variable_scope) is deprecated and will be removed in a future version.
#Instructions for updating:
#non-resource variables are not supported in the long term


#PlaceHolder: Punto de entrada de datos
#Creamos los puntos de entrada de nuestro grafo:

iX = tf.placeholder("float", shape=[None, x.shape[1]]) #Se define el tipo de valor que ingresará y el tamaño.
iY = tf.placeholder("float", shape=[None])

#Topología de la red neuronal:
nn = [2, 16, 8, 1] #Número de neuronas por capa. (3 capas)
lr = 0.01 #Learning Rate del optimizador.

#Capa 1
#Definimos los parámetros que serán optimizados dentro de la red usando "Variable"
#Además, le decimos que sus valores serán inicializados de manera aleatoria usando "random_normal"
w1 = tf.Variable(tf.random_normal([nn[0], nn[1]]), name="weights_1") #y que tendrá las dimensiones "[nn[0], nn[1]]" para este primer caso.
b1 = tf.Variable(tf.random_normal([nn[1]]), name="bias_1")

#Ahora le indicamos que operaciones va a realizar en la capa 1, siendo esta las operaciones: "XW+b"
#Multiplicamos "iX" con "w1" y le sumamos "b1"
l1 = tf.nn.relu(tf.add(tf.matmul(iX, w1), b1)) #Y todo eso, lo pasamos por la función de activación "relu" o "sigmoide".
#Báscimante sería así: f(XW+b)
#l1 = tf.nn.relu(tf.add(tf.matmul(iX, w1), b1)) #Con "relu"

#Capa 2
w2 = tf.Variable(tf.random_normal([nn[1], nn[2]]), name="weights_2")
b2 = tf.Variable(tf.random_normal([nn[2]]), name="bias_2")

l2 = tf.nn.relu(tf.add(tf.matmul(l1, w2), b2))

#Capa 3
w3 = tf.Variable(tf.random_normal([nn[2], nn[3]]), name="weights_3")
b3 = tf.Variable(tf.random_normal([nn[3]]), name="bias_3")

#Vector de predicciones de Y:
pY = tf.nn.sigmoid(tf.add(tf.matmul(l2, w3), b3))[:, 0] #Esto es un tensor.
```

```python
#Evaluación de las predicciones:
loss = tf.losses.mean_squared_error(pY, iY)
#Error cuadrático medio, nos ayudará a comparar lo que obtenemos con lo que tenemos.

#Creamos el optimizador que entrenará a la red:
optimizer = tf.train.GradientDescentOptimizer(learning_rate=lr).minimize(loss)
#Utilizamos el descenso del gradiente para minimizar el coste, es decir, nuestro error.
```

```python
#Número de pasos para entrenar a la red.
n_steps = 3000

iPY = [] # Aquí guardaremos la evolución de las predicción, para la animación.

with tf.Session() as sess: #Iniciamos una sesión para poder ejecutar el grafo (modelo o red neuronal).
    sess.run(tf.global_variables_initializer()) #Inicializa todos los parámetros de pesos (weights) y vallas (bias) tal como los hemos definido.
    
    for step in range(n_steps):
        #Calculamos los tensores "optimizer", "loss" y "pY" | Proporcionamos de los datos por medio de los placeholders usando "feed_dict":
        _, _loss, _pY = sess.run([optimizer, loss, pY], feed_dict={iX: x, iY: y})
        #Los resultados de cada tensor se guardan en las variables respectivas:
        # _ -> resultado de "optimizer" y se nombró así porque no se usará
        # _loss -> resultado de la función de coste "loss"
        # _pY -> resultado del vector de predicciones "pY"
        
        #Ahora graficamos:
        if step%100 == 0: #Cada 25 iteraciones.
            #Cálculo de exactitud entre el vector real y el vector de predicciones:
            acc = np.mean(np.round(_pY)==y) #Redondea "_pY" y lo compara con "y", luego saca la media.
            
            print("Step", step, "/", n_steps, " - Loss = ", _loss, " - Acc", acc)
            
            # Obtenemos predicciones para cada punto de nuestro mapa de predicción _pX.
            _pY = sess.run(pY, feed_dict={ iX : _pX }).reshape((res, res))

            # Y lo guardamos para visualizar la animación.
            iPY.append(_pY)

#Código para la Animación

ims = []

fig = plt.figure(figsize=(10, 10))

print("--- Generando animación ---")

for fr in range(len(iPY)):
    
    im = plt.pcolormesh(_x0, _x1, iPY[fr], cmap="coolwarm", animated=True)

    # Visualización de la nube de datos.
    plt.scatter(x[y == 0,0], x[y == 0,1], c="skyblue")
    plt.scatter(x[y == 1,0], x[y == 1,1], c="salmon")

    plt.title("Resultado Clasificación")
    plt.tick_params(labelbottom=False, labelleft=False)

    ims.append([im])

HTML(animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=1000).to_html5_video())
```

## Usando Keras

```python
import tensorflow.keras as kr

lr=0.01 #Learning rate
nn=[2, 16, 8, 1] #Topología

#Creamos la estructura que contendrá a nuestro modelo:
model = kr.Sequential()

#Capa 1
model.add(kr.layers.Dense(nn[1], activation="relu"))

#Capa 2
model.add(kr.layers.Dense(nn[2], activation="relu"))

#Capa 3
model.add(kr.layers.Dense(nn[3], activation="sigmoid"))

#Compilamos el modelo:
model.compile(loss="mse", optimizer=kr.optimizers.SGD(lr=lr), metrics=["acc"])

#Y ahora lo entrenamos:
model.fit(x, y, epochs=250)
```
